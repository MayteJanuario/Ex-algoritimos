Aula de union e structs

```python
#[c]#
%%writefile ex11.c
#include <stdio.h>
#include <stdlib.h>

typedef struct aluno{
  char nome[100];
  int ra;
  double nota;
} aluno;

double media_notas(aluno * alunos, int n){
  double m = 0.0;
  for(int i = 0; i < n; i++){
    m += alunos[i].nota;
  }
  m = m/n;

  return m;
}



int main(){
  int n;
  aluno * alunos;

  printf("Digite o número de alunos:\n");
  scanf("%d", &n);
  alunos = malloc(sizeof(aluno)*n);
  for(int i = 0; i < n; i++){
    printf("Digite o nome do %d° aluno:\n", &i);
    scanf("%s", alunos[i].nome);
    printf("Digite o ra do %d° aluno:\n"), &i;
    scanf("%d", &alunos[i].ra);
    printf("Digite a nota do %d° aluno:\n"), &i;
    scanf("%lf", &alunos[i].nota);
  }
  printf("A média de nota dos alunos é %.2lf", media_notas(alunos, n));

  return 0;
}
```

```python
#[c]#
!gcc ex11.c -o ex11
!./ex11
```
